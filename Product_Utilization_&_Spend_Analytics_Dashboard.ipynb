```python
# STEP 1: IMPORT LIBRARIES

import pandas as pd
import numpy as np
```

```python
# STEP 2: GENERATE DATASET

np.random.seed(42)

departments = ['ER', 'ICU', 'Surgery', 'Pharmacy']
products = ['Gloves', 'Syringes', 'IV Kits', 'Masks', 'Bandages']

data = []

for _ in range(1000):
    dept = np.random.choice(departments)
    product = np.random.choice(products)

    usage = np.random.randint(10, 500)

    unit_cost = {
        'Gloves': 5,
        'Syringes': 10,
        'IV Kits': 50,
        'Masks': 2,
        'Bandages': 3
    }[product]

    total_cost = usage * unit_cost

    data.append([dept, product, usage, unit_cost, total_cost])

df = pd.DataFrame(data, columns=['Department', 'Product', 'Usage', 'Unit_Cost', 'Total_Cost'])
```

```python
# STEP 3: IDENTIFY HIGH COST DRIVERS

cost_summary = df.groupby('Product')['Total_Cost'].sum().reset_index()
cost_summary = cost_summary.sort_values(by='Total_Cost', ascending=False)

print("=== Cost Drivers ===")
print(cost_summary)
```

```python
# STEP 4: UTILIZATION ANALYSIS

usage_summary = df.groupby('Product')['Usage'].sum().reset_index()
usage_summary = usage_summary.sort_values(by='Usage', ascending=False)

print("\n=== Usage Trends ===")
print(usage_summary)
```

```python
# STEP 5: DEPARTMENT LEVEL ANALYSIS

dept_summary = df.groupby('Department')['Total_Cost'].sum().reset_index()

print("\n=== Department Spend ===")
print(dept_summary)
```

```python
# STEP 6: IDENTIFY INEFFICIENCIES

df['Cost_Per_Unit'] = df['Total_Cost'] / df['Usage']

high_cost = df[df['Cost_Per_Unit'] > df['Cost_Per_Unit'].mean()]

print("\n=== High Cost Inefficiencies ===")
print(high_cost.head())
```

```python
# STEP 7: EXPORT FILES

df.to_csv('utilization_data.csv', index=False)
cost_summary.to_csv('cost_summary.csv', index=False)
dept_summary.to_csv('department_summary.csv', index=False)

print("\nCSV files exported for dashboarding.")
```
